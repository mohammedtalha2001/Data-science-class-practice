```python
catagorical_values=list(data.columns)
catagorical_values.remove('target')
print(catagorical_values)
```

```python
plt.figure(figsize=(30,30))
for i,col in enumerate(catagorical_values,1):
    plt.subplot(4,4,i)
    sns.barplot(x=f"{col}",y='target',data=data)
    plt.ylabel("possibility of having heart disease")
    plt.xlabel(f"{col}")
```
